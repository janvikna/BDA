```python
import networkx as nx
import pylab as plt
```

```python
# Create blank graph
D=nx.DiGraph()
```

```python
#Feed page link to graph
D.add_weighted_edges_from([('A','B',1),('A','C',1),('C','A',1),('B','C',1)])
```

```python
#Print page rank for each pages
print(nx.pagerank(D))
```

```python
#plot graph
nx.draw(D, with_labels=True)
plt.show()
```

```python
start=time.time()
```

```python
response=requests.get('http://recurship.com/')
```

```python
selector=Selector(response.text)
```

```python
href_links=selector.xpath('//a/@href').getall()
```

```python
image_links=selector.xpath('//img/@src').getall()
```
